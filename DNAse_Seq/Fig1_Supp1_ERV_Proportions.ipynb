```r
bedtools_cols <- c("Chm", "Start", "End", "Gen_PeakID", "Score", "Strand", "Signal_value", "neglog10p", "neglog10q", "PeakSummit", "TE_Chr", "TE_Start", "TE_End", "TE_Subfamily", "TE_Length", "TE_Strand", "TE_Family", "Overlap")
E105_overlap <- read.delim(file = './LTR/E105_DNAse_LTRoverlap.bed', sep = '\t', header = F)
colnames(E105_overlap) <- bedtools_cols
nrow(E105_overlap)
E105_overlap <- E105_overlap %>% distinct(TE_Start, .keep_all = T) %>% filter(Overlap > 20) %>% filter(neglog10p >= 2.30)
E105_overlap
nrow(E105_overlap)
```

```r
malelist <- c("M", "MF")
femalelist <- c("F", "MF")
plotmelt <- reshape2::melt(ERV_Times)
plotmelt$Time <- c(rep("E10.5", 6), rep("E12.5", 12), rep("E14.5", 12))
plotmelt$Sex <- c(rep("MF", 6), rep("M", 6), rep("F",6), rep("M",6), rep("F",6))
plotmelt_male <- filter(plotmelt, Sex %in% malelist)
plotmelt_male
write.table(plotmelt_male, file = 'Fig1_male_rawsource.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
plotmelt_female <- filter(plotmelt, Sex %in% femalelist)
plotmelt_female
write.table(plotmelt_female, file = 'Fig1_female_rawsource.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
```

```r
obj_is_vector = function(x){TRUE}
overtime_male <- ggplot(plotmelt_male, aes(x=Time, y=value, group = Family, color = Family)) + geom_line() + geom_point() + scale_color_brewer(palette = "Dark2") + theme_bw() + coord_cartesian(ylim = c(0,70))
overtime_female <- ggplot(plotmelt_female, aes(x=Time, y=value, group = Family, color = Family)) + geom_line() + geom_point() + scale_color_brewer(palette = "Dark2") + theme_bw() + coord_cartesian(ylim = c(0,70))
overtime_male
overtime_female
```

```r
totalpeaks_E105 <- read.delim(file = './E105_DNAse_peaks.narrowPeak', sep = '\t', header = F)
totalpeaks_E105 <- totalpeaks_E105 %>% filter(V8 >= 2.30)
nrow(totalpeaks_E105)
totalpeaks_E125M <- read.delim(file = './E125_M_DNAse_peaks.narrowPeak', sep = '\t', header = F)
totalpeaks_E125M <- totalpeaks_E125M %>% filter(V8 >= 2.30)
nrow(totalpeaks_E125M)
totalpeaks_E125F <- read.delim(file = './E125_F_DNAse_peaks.narrowPeak', sep = '\t', header = F)
totalpeaks_E125F <- totalpeaks_E125F %>% filter(V8 >= 2.30)
nrow(totalpeaks_E125F)
totalpeaks_E145M <- read.delim(file = './E145_M_DNAse_peaks.narrowPeak', sep = '\t', header = F)
totalpeaks_E145M <- totalpeaks_E145M %>% filter(V8 >= 2.30)
nrow(totalpeaks_E145M)
totalpeaks_E145F <- read.delim(file = './E145_F_DNAse_peaks.narrowPeak', sep = '\t', header = F)
totalpeaks_E145F <- totalpeaks_E145F %>% filter(V8 >= 2.30)
nrow(totalpeaks_E145F)
```

```r
rows_te <- c(nrow(E105_overlap), nrow(E125_M_overlap), nrow(E125_F_overlap), nrow(E145_M_overlap), nrow(E145_F_overlap))
rows_total <- c(nrow(totalpeaks_E105), nrow(totalpeaks_E125M), nrow(totalpeaks_E125F), nrow(totalpeaks_E145M), nrow(totalpeaks_E145F))
times <- c("E10.5", "E12.5 Male", "E12.5 Female", "E14.5 Male", "E14.5 Female")
openprop <- data.frame(times, rows_te, rows_total)
openprop$rows_noTE <- (openprop$rows_total - openprop$rows_te)
openprop$Percent_TE <- (openprop$rows_te/openprop$rows_total)*100
openprop$Percent_noTE <- ((openprop$rows_total-openprop$rows_te)/openprop$rows_total)*100
keep <- c(1,5,6)
openplot <- openprop[,keep]
openpropmelt <- reshape2::melt(openplot)
openpropmelt$sex <- c("MF", "M", "F", "M", "F","MF", "M", "F", "M", "F")
openpropmelt$times <- as.factor(openprop$times)
#openpropmelt
#now make a table to plot by raw peak #
openprop
keep <- c(1,2,4)
openraw <- openprop[,keep]
openraw <- reshape2::melt(openraw)
openraw$sex <- c("MF", "M", "F", "M", "F","MF", "M", "F", "M", "F")
openraw
```

```r
order <- c("E10.5", "E12.5 Male", "E14.5 Male", "E12.5 Female", "E14.5 Female")
bothsex_prop <- ggplot(openpropmelt, aes(fill=variable, y=value, x=factor(times, order), label = (round(value)))) + geom_bar(position="stack", stat="identity") + geom_text(size = 3, position = position_stack(vjust = 0.5)) + xlab(label = "Condition") + ylab(label = "% of Detected Peaks") + theme_bw()
bothsex_prop
bothsex_raw <- ggplot(openraw, aes(fill=variable, y=value, x=factor(times, order), label = (round(value)))) + geom_bar(position="stack", stat="identity") + geom_text(size = 3, position = position_stack(vjust = 0.5)) + xlab(label = "Condition") + ylab(label = "Number of Detected Peaks") + theme_bw()
bothsex_raw
```

```r
#now for plots by sex
openpropmeltmale <- filter(openpropmelt, sex %in% malelist)
openpropmeltfemale <- filter(openpropmelt, sex %in% femalelist)
write.table(openpropmeltmale, file = 'Supp1_openpercentmale.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
write.table(openpropmeltfemale, file = 'Supp1_openpercentfemale.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
openrawmale <- filter(openraw, sex %in% malelist)
openrawfemale <- filter(openraw, sex %in% femalelist)
write.table(openrawmale, file = 'Supp1_openrawmale.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
write.table(openrawfemale, file = 'Supp1_openrawfemale.tsv', sep = '\t', quote = F, row.names = F, col.names = T)
```
